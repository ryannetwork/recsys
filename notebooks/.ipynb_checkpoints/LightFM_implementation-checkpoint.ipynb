## Explain

Explanation: https://www.machinelearningmastery.ru/recommendation-system-part-1-use-of-collaborative-filtering-and-hybrid-collaborative-content-in-6137ba64ad58/

1. Матрица взаимодействия пользователя с элементом определяет взаимодействие между пользователем (клиентом) с товаром (продуктом), это может быть показано в виде рейтингов фильмов, которые проголосовали клиенты. Однако в случае с ОКВЭД мы не можем получить явный рейтинг из исторических данных. В этом случае, неявно, я принимаю во внимание «ранг по % - доле во Входящих платежах» в качестве рейтинга, Если клиент A купил продукт в 70% случаев и это наибоьший ранг, то мы можем сказать, что клиент A оценил продукт B с наибольшим рейтингом 82 (так как у нас всего 82 ОКВЭДА). Вы также можете принять во внимание бинарные рейтинги, где 1 относится к клиенту A, который купил или 0 как никогда не покупал продукт B. Здесь пригодится Дискретизация значения - разбиение на бины. 
2. Пользователь матрица взаимодействия представляет собой совместную фильтрацию вклада в модель, матрица взаимодействия предмет-объект определяет особенности предметов. Элемент к функциям может быть представлен как продукт для его метаданных, таких как категория продукта, подкатегория или даже любые фрагменты информации. Можно каждый таргет обогатить информацией в каком ОКАТО находится ИНН итд. Эта матрица может добавить контент-вклад в модель.

```python
train
```

```python
start_time = time.time()
model=LightFM(no_components=115,learning_rate=0.027,loss='warp')
model.fit(train,epochs=12,num_threads=4)
# with open('saved_model','wb') as f:
#     saved_model={'model':model}
#     pickle.dump(saved_model, f)
auc_train = auc_score(model, train).mean()
auc_test = auc_score(model, test).mean()

print("--- Run time:  {} mins ---".format((time.time() - start_time)/60))
print("Train AUC Score: {}".format(auc_train))
print("Test AUC Score: {}".format(auc_test))
```
